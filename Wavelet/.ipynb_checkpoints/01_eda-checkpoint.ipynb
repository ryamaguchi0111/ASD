```python
class permutation_importance():
    def __init__(self, model, metric):
        self.is_computed = False
        self.n_feat = 0
        self.base_score = 0
        self.model = model
        self.metric = metric
        self.df_result = []
    
    def compute(self, X_valid, y_valid):
        self.n_feat = len(X_valid.columns)
        if self.metric == 'auc':
            y_valid_score = self.model.predict_proba(X_valid)[:, 1]
            fpr, tpr, thresholds = roc_curve(y_valid, y_valid_score)
            self.base_score = auc(fpr, tpr)
        else:
            self.base_score = self.metric(y_valid, self.model.predict(X_valid))
        self.df_result = pd.DataFrame({'feat': X_valid.columns, 
                                       'score': np.zeros(self.n_feat),
                                       'score_diff': np.zeros(self.n_feat)})
        
        # predict
        for i, col in enumerate(X_valid.columns):
            df_perm = X_valid.copy()
            np.random.seed(1)
            df_perm[col] = np.random.permutation(df_perm[col])
            y_valid_pred = self.model.predict(df_perm)
            if self.metric == 'auc':
                y_valid_score = self.model.predict_proba(df_perm)[:, 1]
                fpr, tpr, thresholds = roc_curve(y_valid, y_valid_score)
                score = auc(fpr, tpr)
            else:
                score = self.metric(y_valid, y_valid_pred)
            self.df_result['score'][self.df_result['feat']==col] = score
            self.df_result['score_diff'][self.df_result['feat']==col] = self.base_score - score
        self.is_computed = True
    
    def get_negative_feature(self):
        assert self.is_computed!=False, 'compute メソッドが実行されていません'
        idx = self.df_result['score_diff'] < 0
        return self.df_result.loc[idx, 'feat'].values.tolist()
        
    def get_positive_feature(self):
        assert self.is_computed!=False, 'compute メソッドが実行されていません'
        idx = self.df_result['score_diff'] > 0
        return self.df_result.loc[idx, 'feat'].values.tolist()
    
    def show_permutation_importance(self, score_type='loss'):
        '''score_type = 'loss' or 'accuracy'  '''
        assert self.is_computed!=False, 'compute メソッドが実行されていません'
        if score_type=='loss':
            ascending = True
        elif score_type=='accuracy':
            ascending = False
        else:
            ascending = ''
        
        plt.figure(figsize=(15, int(0.25*self.n_feat)))
        sns.barplot(x="score_diff", y="feat", data=self.df_result.sort_values(by="score_diff", ascending=ascending))
        plt.title('base_score - permutation_score')
```

```python
def get_spectrum_df(DATA_DIR):
    files = glob.glob(DATA_DIR + 'spectrum_raw/*')
    spec_list = []
    i = 0
    for file in files:
        spectrum_df = pd.read_csv(file, sep='\t', header=None)
        spectrum_df.columns = ['wavelength', 'intensity']
        spectrum_df['spectrum_filename'] = file.split(sep='/')[-1]
        spec_list.append(spectrum_df)

    return pd.concat(spec_list, axis=0)
```

```python
def get_statics(df, feature_names, group_name):
    mean_feature_names = [f + '_mean' for f in feature_names]
    std_feature_names = [f + '_std' for f in feature_names]
    abssum_feature_names = [f + '_abssum' for f in feature_names]
    max_feature_names = [f + '_max' for f in feature_names]
    min_feature_names = [f + '_min' for f in feature_names]
    
    df_mean = df.groupby(group_name).mean()
    df_mean = df_mean[feature_names]
    df_mean.columns = mean_feature_names

    df_std = df.groupby(group_name).std()
    df_std = df_std[feature_names]
    df_std.columns = std_feature_names

    df_abssum = df[feature_names].abs()
    df_abssum = pd.concat([df_abssum, df[group_name]], axis=1)
    df_abssum = df_abssum.groupby(group_name).sum()
    df_abssum.columns = abssum_feature_names
    
    df_max = df.groupby(group_name).max()
    df_max = df_max[feature_names]
    df_max.columns = max_feature_names
    
    df_min = df.groupby(group_name).min()
    df_min = df_min[feature_names]
    df_min.columns = min_feature_names
    
    df_statics = pd.merge(df_mean, df_std, on=group_name)
    df_statics = pd.merge(df_statics, df_abssum, on=group_name)
    df_statics = pd.merge(df_statics, df_max, on=group_name)
    df_statics = pd.merge(df_statics, df_min, on=group_name)
    
    return df_statics
```

### predict

```python
params = {
    'eta': 0.1,
    'max_depth': 5,
    'colsample_bytree': 0.8,
    'subsample': 0.8,
    'gamma': 0,
    'lambda': 1,
    'alpha': 0,
    'min_child_weight': 1,
    'n_estimators': 400
}
```

```python
_n_splits = 10

scores = []
precision_scores = []
f1scores = []
models = []
permutation_importances = []
conf_matrix = np.zeros((2, 2))
_X = df_train_fit.drop(not_feature_list, axis=1)
_y = df_train_fit['target']
skf = StratifiedKFold(n_splits=_n_splits, shuffle=False, random_state=RANDOM_STATE)
for train_idx, valid_idx in skf.split(_X, _y):
    X_train, y_train = _X.iloc[train_idx], _y.iloc[train_idx]
    X_valid, y_valid = _X.iloc[valid_idx], _y.iloc[valid_idx]
    model = xgb.XGBClassifier()
    model.set_params(**params)
    model.fit(X_train, y_train)
    y_valid_predict = model.predict(X_valid)
    score = accuracy_score(y_valid, y_valid_predict)
    conf_matrix = conf_matrix + confusion_matrix(y_valid, y_valid_predict)
    precision_score = average_precision_score(y_valid, y_valid_predict)
    f1score = f1_score(y_valid, y_valid_predict)
    scores.append(score)
    models.append(model)
    precision_scores.append(precision_score)
    f1scores.append(f1score)
    pi = permutation_importance(model, accuracy_score)
    pi.compute(X_valid, y_valid)
    permutation_importances.append(pi)
```

```python
pi_score_mean = np.mean([permutation_importances[i].df_result['score_diff'] for i in range(_n_splits)], axis=0)
pi_score_var  = np.var([permutation_importances[i].df_result['score_diff'] for i in range(_n_splits)], axis=0)
df_pi = pd.DataFrame({
    'feat': permutation_importances[0].df_result['feat'],
    'score_diff_mean': pi_score_mean,
    'score_diff_var': pi_score_var
})
plt.figure(figsize=(15, int(0.25*df_pi.shape[0])))
sns.barplot(x="score_diff_mean", y="feat", data=df_pi.sort_values(by='score_diff_mean', ascending=False))
plt.title('base_score - permutation_score')
```
